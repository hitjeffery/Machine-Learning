#### 非线性SVM回归

**数据生成**
+ x：在区间$[-3,3]$之间均匀分布
+ y_pure：真实预测数据
$$ y = \sin(4x)+x $$
+ y：添加噪声数据，噪声分布$N(0,0.25)$
$$ y = \sin(4x)+x+\delta,\qquad \delta \sim N(0,0.25) $$

```python
import numpy as np
import matplotlib.pyplot as plt

rnd = np.random.RandomState(42)
x = rnd.uniform(-3,3,size=100)
y_pure = np.sin(4*x) + x
y = y_pure + rnd.normal(size=len(x))/4

plt.plot(x,y,'+')

xx = np.linspace(-3, 3, 1000).reshape(-1, 1)
yy = np.sin(4*xx) + xx
plt.plot(xx,yy)

plt.show()
```

**学习非线性SVM回归器**

函数SVR学习，使用模型参数
+ 高斯核函数
+ 核函数参数：$\gamma = 1$
+ 调和参数：$C=100$
+ 误差参数：$\epsilon=0.01$

```python
from sklearn.svm import SVR

svr = SVR(kernel='rbf',gamma=1,C=100,epsilon=0.01).fit(x.reshape(-1,1), y)
predict_y = svr.predict(xx)

plt.plot(x,y,'+')
plt.plot(xx,yy)

plt.plot(xx,predict_y,'b')

plt.show()
```
