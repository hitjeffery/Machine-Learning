#### 神经网络分类

**数据生成**
+ 生成100个二分类2维数据
+ 两类数据分布在交叠的两个半圆形区域
+ 数据集划分为训练集和测试集

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=100, noise=0.25, random_state=3)
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y,random_state=42)

plt.scatter(X[:,0], X[:,1], c=y, s=50, cmap='rainbow')

plt.show()
```

**学习神经网络
+ 网络结构
    - 输入层：2个神经元
    - 隐含层：6个神经元，对数Sigmoid激活函数
    - 输出层：1个神经元，对数Sigmoid激活函数
+ 学习参数
    - 学习算法：lbfgs
    - 最大迭代次数：40

```python
from sklearn.neural_network import MLPClassifier
from plot_decision_boundary import plot_decision_boundary

eps = 0.5
x_min,x_max = X[:,0].min()-eps, X[:,0].max()+eps
y_min,y_max = X[:,1].min()-eps, X[:,1].max()+eps

fig, axes = plt.subplots(2,4,figsize=(18, 6))
for iter,ax in zip([1,5,10,40,80,100,200,500],axes.reshape([8,1])):
    mlp = MLPClassifier(hidden_layer_sizes=(6), solver='lbfgs', activation='logistic', max_iter=iter, random_state=0).fit(X_train, y_train)
    plot_decision_boundary(mlp,axis=[x_min,x_max,y_min,y_max],ax=ax[0])
    ax[0].scatter(X[:,0], X[:,1], c=y, s=50, cmap='rainbow')
    ax[0].set_title('iter=%d,train=%4.3f,test=%4.3f' %(iter,mlp.score(X_train,y_train),mlp.score(X_test,y_test)))

plt.show()
```
