#### 初始簇均值对k-means算法的影响

+ 使用西瓜的密度和含糖量数据
+ 随机选择3个聚类数据，作为初始的簇均值
+ k-means算法聚类为3个簇，显示聚类结果
+ 计算聚类结果的SSD(Sum of Squared Distances)：
$$ \text{SSD} = \sum_{i=1}^k{\sum_{\mathbf{x}\in C_i}{\|\mathbf{x}-\boldsymbol{\mu}_i \|^2_2}} $$
+ $\boldsymbol{\mu}_i$为簇$C_i$的均值向量

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

X = np.array([[0.697,0.460],[0.774,0.376],[0.634,0.264],[0.608,0.318],[0.556,0.215],[0.403,0.237],[0.481,0.149],[0.437,0.211],[0.666,0.091],[0.243,0.267],
              [0.245,0.057],[0.343,0.099],[0.639,0.161],[0.657,0.198],[0.360,0.370],[0.593,0.042],[0.719,0.103],[0.359,0.188],[0.339,0.241],[0.282,0.257],
              [0.748,0.232],[0.714,0.346],[0.483,0.312],[0.478,0.437],[0.525,0.369],[0.751,0.489],[0.532,0.472],[0.473,0.376],[0.725,0.445],[0.446,0.459]])

fig, axes = plt.subplots(2,4,figsize=(25, 8))
for seed, ax in zip(range(8),axes.reshape([8,1])):
    kmeans = KMeans(n_clusters=3, n_init=1, random_state=seed)
    Labels = kmeans.fit_predict(X)

    ax[0].scatter(X[:,0],X[:,1],c=Labels,cmap='rainbow')
    ax[0].set_title("SSD=%f" %(kmeans.inertia_))

plt.show()
```

**尝试多个初始值**

+ 随机尝试10组不同的初始簇均值设置
+ 计算每一组聚类结果的SSD
+ 选择SSD最小的一次结果作为k-means的聚类结果

```python
kmeans = KMeans(n_clusters=3, n_init=10)
Labels = kmeans.fit_predict(X)

plt.scatter(X[:,0],X[:,1],c=Labels,cmap='rainbow')
plt.title("SSD=%f" %(kmeans.inertia_))

plt.show()
```
