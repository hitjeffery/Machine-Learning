#### 非线性流形学习

**数据生成**
+ 随机生成1000个3维数据，均匀分布在一个S型的曲面上
+ 显示3维数据

```python
import matplotlib.pyplot as plt
from sklearn import manifold, datasets

X, color = datasets.make_s_curve(1000, random_state=0)

fig = plt.figure(figsize=[10,8])
ax = plt.axes(projection="3d")
ax.scatter(X[:,0], X[:,1], X[:,2], s=50,c=color, cmap=plt.cm.Spectral)
ax.view_init(15, -72)

plt.show()
```

**流形学习**

+ 分别使用PCA算法、MDS算法和Isomap算法将数据映射到2维空间
+ 显示3种算法得到的2维数据分布

```python
from sklearn.decomposition import PCA
from sklearn.manifold import MDS, Isomap

figure, axes = plt.subplots(1, 3, figsize=(20, 3))

Y = PCA(n_components=2).fit_transform(X)
axes[0].scatter(Y[:, 0], Y[:, 1], c=color, cmap=plt.cm.Spectral)
axes[0].set_title('PCA')

Y = MDS(n_components=2, max_iter=100, n_init=1).fit_transform(X)
axes[1].scatter(Y[:, 0], Y[:, 1], c=color, cmap=plt.cm.Spectral)
axes[1].set_title('MDS')

Y = Isomap(n_neighbors=10, n_components=2).fit_transform(X)
axes[2].scatter(Y[:, 0], Y[:, 1], c=color, cmap=plt.cm.Spectral)
axes[2].set_title('Isomap')

plt.show()
```
