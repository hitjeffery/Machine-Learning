#### 参数对DBSCAN算法的影响

+ 随机生成12个2维数据
+ 设置参数：$\text{MinPts} = \{2,3,5\},\epsilon = \{1,1,5,2\}$
+ 分别设置两个参数的12种组合，使用DBSCAN算法聚类仿真数据
+ 显示聚类结果

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs

X,y = make_blobs(random_state=0, n_samples=12)

fig, axes = plt.subplots(3,4,figsize=(20, 10))
for i, MinPts in enumerate([2, 3, 5]):
    for j, epsilon in enumerate([1, 1.5, 2, 3]):
        dbscan = DBSCAN(min_samples=MinPts, eps=epsilon)
        clusters = dbscan.fit_predict(X)

        print('MinPts=%d epsilon=%f  cluster: %s' % (MinPts, epsilon, clusters))
        id = np.where(clusters==-1)
        axes[i,j].scatter(X[id,0], X[id,1], marker="o", c="white", edgecolors="black",s=30)
        id = np.where(clusters!=-1)
        axes[i,j].scatter(X[id,0], X[id,1], c=clusters[id], s=50, cmap="rainbow")
        axes[i,j].set_title(r'$MinPts=%d, \epsilon=%2.1f$' %(MinPts,epsilon))

plt.plot()
```
