#### MNIST数据集

+ 手写数字数据集
+ 来自美国国家标准与技术研究所(NIST: National Institute of Standards and Technology)
+ 数字0~9，共10个类别
+ 训练数据：60000
+ 测试数据：10000
+ 数据格式：$28\times 28$的灰度图像，像素取值0~255

```python
import numpy as np
import pandas as pd

data = pd.read_csv("MNIST_train.csv")
X_train = data.iloc[:,1:785].to_numpy()
y_train = data.iloc[:,0].to_numpy()

data = pd.read_csv("MNIST_test.csv")
X_test = data.iloc[:,1:785].to_numpy()
y_test = data.iloc[:,0].to_numpy()

print("Train Samples:", X_train.shape)
print("Test Samples:", X_test.shape)
print("Train Labels:", y_train[0:10])
```

**显示前100个训练样本图像**

```python
import matplotlib.pyplot as plt

fix, axes = plt.subplots(5, 20, figsize=(20, 5),subplot_kw={'xticks': (), 'yticks': ()})
for x, ax in zip(X_train, axes.ravel()):
    image = x.reshape([28,28])
    ax.imshow(image, cmap="gray")

plt.show()
```
